```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
marriage_data = pd.read_csv("us-marriages-divorces-1867-2014.csv")
```

```python
marriage_range=marriage_data[(marriage_data['Year']>1867) & (marriage_data['Year']<2014) ]
marriage_range
```

```python
date_marriage_work=marriage_range[['Year','Marriages']]
date_marriage_work.set_index('Year',inplace=True)
date_marriage_work
```

```python
marriage_divorced=marriage_range[['Year','Divorces']]
marriage_divorced.set_index('Year',inplace=True)
marriage_divorced
```

```python
data_main=pd.concat([date_marriage_work,marriage_divorced],axis=1)
data_main.reset_index(inplace=True)
data_main
```

```python
plt.plot('Year','Marriages',data=data_main)
plt.plot('Year','Divorces',data=data_main)
plt.title("Marriage and Divorces per Year")
plt.ylabel("Count")
plt.xlabel("Year")
plt.legend()
plt.show()
```
